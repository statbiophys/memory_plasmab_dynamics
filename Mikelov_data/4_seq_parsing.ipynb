```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from os import listdir
```

Here we create the final metadata file, we keep only the necessary information from the changeO pipeline and we organize the samples in replicates and merged replicates

```python
metadata_samp = pd.read_csv('metadata/metadata_samples.tsv', sep='\t')
time_id = np.array(metadata_samp['time'], dtype='str')
cell = np.array(metadata_samp['cell'], dtype='str')
metadata_samp['id'] = metadata_samp['patient'] + '_t' + time_id + '_' + cell
agg_dict = {'patient':'first', 'age':'first', 'time':'first', 'cell':'first',
           'replicate' : list, 'sample_n' : list}
metadata = metadata_samp.groupby('id').agg(agg_dict)
metadata['n_repls'] = metadata['replicate'].str.len()
```

```python
metadata
```

```python
def replicate_preprocess(fr):
    """
    Discarding unproductive seqs, filters V and J calls,
    and keeps only a subsets of fields.
    """
    
    fr = fr[fr.productive == 'T']
    
    # Keeping only the top-score V and J genes
    fr.loc[:,'v_call'] = fr.v_call.str.split(',').str[0]
    fr.loc[:,'j_call'] = fr.j_call.str.split(',').str[0]
    
    return fr[['sequence', 'v_call', 'j_call', 'c_call', 'sequence_alignment', 'germline_alignment', 
            'junction', 'junction_aa', 'junction_length', 'v_sequence_start',
            'v_sequence_end', 'v_germline_start', 'v_germline_end', 'j_sequence_start', 
            'j_sequence_end', 'j_germline_start', 'j_germline_end']]
```

```python
sample = '1'
fr = pd.read_csv('changeO_out/' + str(sample) + '_igblast_db-pass.tsv', sep='\t', index_col=0)
fr = replicate_preprocess(fr)
```

```python
# Merging replicates and saving the full sample
# Saving the single replicate when more than 1 replicate is present
    
for _id, row in metadata.iterrows():

    merged_fr = pd.DataFrame()
    for ir, sample in enumerate(row.sample_n):

        print(sample)
        fr = pd.read_csv('changeO_out/' + str(sample) + '_igblast_db-pass.tsv', sep='\t', index_col=0)
        fr = replicate_preprocess(fr)
        merged_fr = pd.concat((merged_fr, fr))
        if len(row.sample_n):
            fr.to_csv('sequences/replicates/' + _id + '_r' + str(ir + 1) + '.tsv', sep='\t')

    #agg_dict = {k : 'first' for k in merged_fr.columns}
    #agg_dict['count'] = sum
    #merged_fr = merged_fr.groupby('sequence').agg(agg_dict)
    merged_fr.to_csv('sequences/' + _id + '.tsv', sep='\t')
```
