```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Visualizing the count histograms of each sample

```python
metadata = pd.read_csv('metadata/metadata.tsv', sep='\t', index_col=0)
metadata
```

```python
count_pairs_repl_fr = pd.DataFrame(index=metadata.index, columns=[3,1,2])
for s_name in metadata.index:
    
    if metadata.loc[s_name].n_repls == 1:
        f = pd.read_csv('sequences/'+s_name+'.tsv', sep='\t', index_col=0)
        count_pairs_repl_fr.loc[s_name, 1] = len(f)
        continue
        
    for i in range(metadata.loc[s_name].n_repls):
        r_id = s_name + '_r' + str(i+1)
        f = pd.read_csv('sequences/replicates/'+r_id+'.tsv', sep='\t', index_col=0)
        count_pairs_repl_fr.loc[s_name, i+1] = len(f)
```

```python
m_indexes = metadata[metadata.cell == 'mem'].index
p_indexes = metadata[metadata.cell == 'pb'].index
```

```python
count_pairs_repl_fr = count_pairs_repl_fr.fillna(0)
count_pairs_repl_fr.loc[m_indexes].plot(kind="bar",figsize=(6, 3), width=0.9)
plt.ylabel('N paired cells', fontsize=12)
plt.ylim(0, 60000)
plt.tight_layout()
plt.savefig('plots/N_seqs_replicates_m.svg', dpi=200)
```

```python
count_pairs_repl_fr = count_pairs_repl_fr.fillna(0)
count_pairs_repl_fr.loc[p_indexes].plot(kind="bar",figsize=(6, 3), width=0.9)
plt.ylabel('N paired cells', fontsize=12)
#plt.ylim(0, 160000)
plt.tight_layout()
plt.savefig('plots/N_seqs_replicates_p.svg', dpi=200)
```
